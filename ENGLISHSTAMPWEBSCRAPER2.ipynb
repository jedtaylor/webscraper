```python
nametag = "div"
namenested = "h1"
nameclass = "title"

sizetag = "div"
sizeclass = "product--info"

pricetag = "div"
priceclass = "content product--details"


testlink = 'https://www.englishstamp.com/p/4214/modern-wedding-save-the-date-stamp'

productlist = []



def getname(soup, nametag, namenested, nameclass, nestedtag=None):
    name_element = soup.find(nametag, class_=nameclass)
    if nestedtag:
        name = name_element.find(namenested).text.strip()
    else:
        name = name_element.text.strip()
    return name


def getsize(soup, sizetag, sizeclass):
    try:
        size_pattern = r'(\d+mm)'  
        size = re.search(size_pattern, str(soup)).group(1) 
    except:
        pass    
    try:
        size_pattern = r'(\d+mm x \d+mm|\d+mm X \d+mm|\d+mm × \d+mm)'  
        size = re.search(size_pattern, str(soup)).group(1)  
    except:
            print('not listed')
    return size


def getprice(soup, pricetag, priceclass):

    try:
        price_pattern = r'(£\d+\.\d{2})'  
        price = re.search(price_pattern, str(soup)).group(1)     

        pricelen = len(price)
        if (pricelen > 5):
            price_pattern = r'(£\d+\.\d{2})'  
            price = re.search(price_pattern, str(soup)).group(1)  
            saleprice = soup.find(pricetag, class_=priceclass).text.split()

        else:
            price_pattern = r'(£\d+\.\d{2})'  
            price = re.search(price_pattern, str(soup)).group(1)  
            saleprice = ('not on sale')

    except:
        saleprice = ('not on sale')
        pass
            
    return price, saleprice


try:

    #for link in linkholder:
    for link in range(0,1):       
        #r = rq.get(link, headers=headers)
        r = rq.get(testlink, headers=headers)
        soup = BeautifulSoup(r.content, 'lxml')

        name = getname(soup, nametag, namenested, nameclass)
        size = getsize(soup, sizetag, sizeclass)
        price, saleprice = getprice(soup, pricetag, priceclass)
        
        products = {

                    'name': name,
                    'size': size,
                    'price': price,
                    'saleprice': saleprice

                }

        productlist.append(products)

except:
    print('URL not recognised') 

productdf = pd.DataFrame(productlist)

#productdf['name'] = productdf['name'].astype(str)
#productdf['size'] = productdf['size'].astype(str)
#productdf['price'] = productdf['price'].astype(str)
#productdf['saleprice'] = productdf['saleprice'].astype(str)


productdf.drop_duplicates(inplace=True)

print (len(productdf))
productdf.head(10)
```
